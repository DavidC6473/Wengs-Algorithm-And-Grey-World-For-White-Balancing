```python
def Weng(img):
    
    #RGB to YCbCr
    YCbCr = color.rgb2ycbcr(img)
    
    Y   = YCbCr[:,:,0]
    Cb  = YCbCr[:,:,1]
    Cr  = YCbCr[:,:,2]

    # Mean average values of Cb & Cr
    Mb = np.mean(Cb)
    Mr = np.mean(Cr)
    
    # Average absolute differences for the components
    Db = np.mean(np.absolute(Cb - Mb))
    Dr = np.mean(np.absolute(Cr - Mr))
    
    # Classifying near-white pixels
    # |Cb(i,j) - (Mb + Db * sgn(Mb))| < 1.5 * Db
    
    # |Cb(i,j) - (Mb + Db * sgn(Mb))|
    Ab = np.absolute(Cb - (Mb + (Db * np.sign(Mb)))) 
    
    # |Cr(i,j) - (Mr + Dr * sgn(Mr))| < 1.5 * Dr
    Ar = np.absolute(Cr - (Mr + (Dr * np.sign(Mr)))) 
    
    # make pixels in Ab < ADb = 0
    ADb = 1.5 * Db
    Ab[Ab > ADb] = 0
    
    # Make pixels in Ar < ADr = 0    
    ADr = 1.5 * Dr
    Ar[Ar > ADr] = 0
    
    # All pixels that satisfy both conditions are identified
    nw = np.multiply(Ar , Ab)
    
    # store all values above 0
    nw_pix = Y[nw > 0]
    
    # near-white in the top 10 percentile
    Y_90 = np.percentile(Y[nw > 0], 90)
    
    # R G & B Value
    R  = img[:,:,0]
    G  = img[:,:,1]
    B  = img[:,:,2]
    
    R_Y = R[Y >= Y_90]
    G_Y = G[Y >= Y_90]
    B_Y = B[Y >= Y_90]
    
    # Ravg
    Ravg = R_Y.mean()
    
    # Gavg
    Gavg = G_Y.mean()
    
    # Bavg
    Bavg = B_Y.mean()
    
    # Ymax
    Ymax = Y.max()
    
    # α
    α = Ymax/Ravg
    
    # γ
    γ = Ymax/Gavg
    
    # β
    β = Ymax/Bavg
    
    # α * R
    Radj = np.clip((α * R), 0, 255).astype('uint8')
    
    print("R")
    print(R)
    print("Radj")
    print(Radj)
    
    # γ * G
    Gadj = np.clip((γ * G), 0, 255).astype('uint8')
    print("G")
    print(G)
    print("Gadj")
    print(Gadj)
    
    # β * B
    Badj = np.clip((β * B), 0, 255).astype('uint8')
    print("B")
    print(B)
    print("Badj")
    print(Badj)
    
    print("---------------------")
    print("---------------------")
    print("---------------------")
    return np.dstack((Radj, Gadj, Badj))
```
